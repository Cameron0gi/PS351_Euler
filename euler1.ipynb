```python
""""
Set time step: dt = 0.1
Set maximum time (tmax): e.g. 10
Set value for tau: e.g. 2.0
Set initial conditions: t = 0, N(0) = e.g. 10
Start loop
N(t+dt) = N(t) - N(t)/tau dt (Calculate new N value)
t = t + dt (Calculate next time)
print N and t to file (Print new values to a file)
loop until ABS(t - tmax) < dt/2 (loop exit condition)
"""
import math

dt=0.01
tmax=10
tau=2.0

#initial coditions
t=0
n=10

while(abs(t-tmax) > (dt/2)):
    with open('outputdt1.txt', 'a') as f:
        f.write(str(t)+'\n')
        f.write(str(n)+'\n')
    n=n-((n*dt)/tau)
    t=t+dt
    
#calclate t using eqn2
t=0
n=10

while(abs(t-tmax) > (dt/2)):
    with open('outputcalc1.txt', 'a') as f:
        f.write(str(t)+'\n')
        f.write(str(n)+'\n')
    n=n*math.exp(-t/tau)
    t=t+dt


    
```
